#### Method
<a id ="method"></a>

- Function jo class ke andar define hota hai use method kehte hain.
- Function that are biuld inside the class are called method
- there are three different type of method
    - Object Method: 
        - It works with object data (instance variables) and uses **self**.
        - through object method change made in lcass attrubte onpy affect that object onyl and other object of the class 
        - If you change a class attribute using an object method , it creates a new attribute only for that object, so other objects are not affected.
    - Class Method:
        - It works with class variables and uses **cls**.
        - Defined using **@classmethod** decorator.
        - If a class attribute is changed using a class method, the change affects the entire class, so all objects see the updated value.
    - Static Method:
        - It does not use class or object data.
        - Defined using **@staticmethod** decorator.
        
    

````python
### Object Method
class Student:
    coll = 'JECRC'
    loc = 'Jaipur'
    Depr = 'CSE'
    
    def __init__(self,name,age,roll_no,DOB,batch):
        self.name = name
        self.age = age
        self.roll_no = roll_no
        self.DOB = DOB
        self.batch = batch
    
    
    ## Object method because class attrubuted can also be called
    # def display(self, loc):
    def display(self):
        # self.loc = loc
        print(self.name, self.age, self.roll_no, self.DOB, self.loc)
    
    # def display(self):
    #     print(f"Your Name is: {self.name}")
    #     print(f"Your age is: {self.age}")
    #     print(f"Your roll No: {self.roll_no}")
    #     print(f"Your DOB: {self.DOB}")
    #     print(f"Your batch: {self.batch}")
    #     print(f"YOur Location: {self.loc}")
        
        
    ### class method 
    
    @classmethod
    def show_college(cls):
    # def show_college(cls, coll):
        # cls.coll = coll
        print(f"College: {cls.coll}")
        print(f"Location: {cls.loc}")
        print(f"Department: {cls.Depr}")
        
    ## Static method
    @staticmethod
    def greet():
        print("Welcome to Student Management System")
    
        
st4 = Student('Aditya', 232, 'qwer','58/85/8520', 'cse')
st5 = Student('samm', 766, 'qzxc','36/36/5520', 'cre')

# st4.display("sikar")
# print(st5.loc)

# Student.show_college("pornima")
# st4.show_college("skit")
st5.coll
# Student.show_college()
# st4.show_college()

Student.greet()
st4.greet()
````

#### Abstraction
<a id ="abstraction"></a>

- Hiding internal implementation details and showing only necessary features.
- Hiding internal details

````python
# Abstraction example

class Car:
    
    def __init__(self):
        self.acc = False
        self.clutch = False
        self.brk = False
        
    def start(self):
        self.acc = True
        self.clutch = True
        print("Car started.....")
        
c1 = Car()
c1.start()
````
