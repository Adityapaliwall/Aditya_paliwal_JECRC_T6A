#### Gloabl and Local Variable
<a id ="globalvar"></a>

```python
# Global variable
x = 10
y = 100

def demo():
    # Local variable
    x = 20
    print("Local x inside function:", x)   # uses local x

    # Access global variable
    print("Access global y inside function:", y)

def modify():
    global x   # modify global variable
    x = 50
    print("Modified global x inside function:", x)

print("Initial global x:", x)

demo()        # shows local vs global access

modify()      # edits global variable

print("Final global x after modification:", x)
```

```python
def prodl(li):
    mul = 1
    for i in li:
        mul *= i
    return mul

prodl([1,2,3,4,5,6])
```

```python
## write a program to print a initila index of the character present in a given string
def char_index(string, char):
    for i in range(len(string)):
        if string[i] == char:
            return i
    return -1

char_index(eval(input("Enter the string: ")), eval(input("Enter the char: ")))


```
