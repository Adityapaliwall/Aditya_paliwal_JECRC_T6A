#### example of oops


````python
### oops example
class Bank:
    
    def __init__(self,name,phone_no,balance):
        self.name = name
        self.phone_no = phone_no
        self.balance = balance
        
    
    def deposit(self,amount):
        if amount < 0:
            print("Invalid amount")
            return
            
        self.balance += amount
        print(f"Deposit successful. New balance: {self.balance}")
    
    def withdraw(self,amount):
        if amount < 0:
            print("Invalid amount")
            return
            
        if amount > self.balance:
            print("Insufficient balance")
            return
            
        self.balance -= amount
        print(f"Withdrawal successful. Amount deducted : {amount}")
        print(f"Withdrawal successful. New balance: {self.balance}")
        
        
    def check_balance(self):
        print(f"Current balance: {self.balance}")
        
    
user1 = Bank("Aditya",9174809975,100)
user1.check_balance()
user1.deposit(400)
user1.withdraw(1000)
user1.withdraw(100)
user1.check_balance()
    
        
````

#### Inheritance
<a id ="inheritance"></a>

- Inheritance means a child class gets the properties and methods of a parent class.
- It helps in reusing code so we don’t need to write the same code again.
- here  
    - Parent → base class
    - Child → derived class (inherits from Parent)
- Type of Inheritance
    - Single Inheritance
        - One child class inherits from one parent class.

    - Multiple Inheritance
        - One child class inherits from more than one parent class.

    - Multilevel Inheritance
        - A class inherits from another child class.

    - Hierarchical Inheritance
        - Multiple child classes inherit from one parent class.
- Property Method
    - The @property decorator is used to access a method like an attribute.

#### Polymorphism 
<a id = "polymorphism"></a>

- Polymorphism means:
- Same operation behaves differently depending on context.

````python
class CreditCard:
    
    def pay(self, amt):
        print(f"Paid {amt} using Credit Card")


class UPI:
    
    def pay(self, amt):
        print(f"Paid {amt} using UPI")


class Cash:
    
    def pay(self, amt):
        print(f"Paid {amt} using Cash")


payments = [CreditCard(), UPI(), Cash()]

for p in payments:
    p.pay(500)
````

#### File_Handling
<a id ="filehandling"></a>

- File handling is used to create, read, write, append, and manage files in Python.
- File is a type of container in which we contain or store some data.  
  by using extension name, we can identify what type of data is there inside of it   
  ex:- .txt, .py, .html ,,,,etc
- Before handling any file, taking permission is very much important,  
- **open():**  
    **open('filename.ext'/'absolute_path',mode)**
- **close():**  
    **var_name.close()**

- there are 3 different modes are present
    - write(w)
    - read(r)
    - append(a)
1. **Write mode:**
    - only write **(w)**
    - write + read **(w+)**
    - write binary **(wb)**
    - write & read binary **(wb+)**

2. **Read mode:**
    - only read **(r)**
    - read + write **(r+)**
    - read binary **(rb)**
    - read & write binary **(rb+)**

3. **Append  mode:**
    - only append **(a)**
    - append + read **(a+)**
    - append binary **(ab)**
    - append & read binary **(ab+)**

- For write operations:
    - in this , if the file is not present, it will create one then perform write operation.
    - if the file is already present, them it will override the pre data with the new one.
    1. **write(str_data):** Single line of data
    2. **writelines([lin11,line2.....]) :** multiple line of data

- For wirte Operations:
    1. **read():** Display whole file in normal format
    2. **readline():** Reads only one line from the file.
    3. **readlines():** Reads all lines and returns them as list.



````python
## code of read and write

file = open('temp.txt','w')
file.write("heloo how are you \n")
file.writelines(["how are you\n", "goku\n", "anime\n", "pigeon\n"])

file.close()

````
