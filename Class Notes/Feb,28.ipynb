#### Functions
<a id = "functions"></a>

- A function is a reusable block of code that performs a specific task.
- there are two types of functions
    - In built functions
        - Utility In-Built Function --> type, id, hex, len, dir
        - String In-Built Function --> isupper, islower, swapcase, replace, capitalize, title, split, join, index, count, upper,          lower, isdigit, strip,  rstrip,
        - List In-Built Function --> sort, append, remove, pop, insert, count, index, reverse, 
        - Tuple In-Built Function --> index, count
        - Set In-Built Function --> add, remove, pop, update, union, intersection, difference, clear
        - Dict In-Built Function --> get, items, values, keys, pop, popitem, update, clear
    - user defined function
        - Syntax --> **def f_name():**
        - Types of UD_Function:
            - Function without return values & without arguments.
            - Function without return values with arguments.
            - Function with return values , without argument.
            - Function with return values , with arguments.
