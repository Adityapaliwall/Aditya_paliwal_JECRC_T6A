#### Type of Arguments
<a id="arguments"></a>

- Positional Arguments --> 
    - Arguments are passed in the same order as parameters in the function.
    - Values are passed to parameters in the same order as defined in the function.
- Default Arguments --> 
    - Parameters have default values if no argument is passed.
    - Parameters have predefined default values used when no value is provided during the function call.
- Keyword Arguments -->
    - Values are passed using parameter names, so order does not matter.
    - Arguments are passed using parameter names.
- Variable Length Arguments (*args) -->
    - Allows multiple positional arguments.
    - Allows passing multiple positional arguments, which are stored as a tuple.
    - Allows passing multiple keyword arguments, which are stored as a dictionary.

- ***1️⃣ Positional Arguments → 2️⃣ Default Arguments → 3️⃣ Variable Positional Arguments (*args) → 4️⃣ Keyword Arguments***

```python
## with all the argumens

def form(name, mail, age=18, *skills, phone):
    print("Name is:", name)
    print("Mail is:", mail)
    print("Phone No:", phone)
    print("Age is:", age)
    print("Skills:", skills)


# Function call using different argument types
form("Aditya", "aditya@paliwal", 9174807792, "Python", "Java", "ML", phone=22)
```
