```python
## Question 1 
## Fee Calculator
class Solution:
    def parking_fee(self, hours):
        fee = 0
        #write your code here
        if hours <=2:
            fee = 0
        elif hours <= 5:
            fee = (hours-2)*20
        else:
            fee = (3*20) + (hours - 5)*50
        if hours > 12:
            fee += 200
        
        return fee
```

```python
## Question 2
##Result Analyzer
class Solution:
    def quiz_score(self, student, correct):
        score = 0
        #write your code here
        for i in range(len(student)):
            if student[i] =='X':
                score +=0
            elif student[i] == correct[i]:
                score += 5
            else:
                score -= 2

        to_po = len(correct) * 5
        
        if score >0.7*to_po:
            score += 10
        return score
```
