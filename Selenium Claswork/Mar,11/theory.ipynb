
- Software testing - Checking if software works as per user requirements.  
2 Ways of testing ->  
    1. Manual   
        ~ Disadvantage ->   
            a. Time consuming  
            b. Tedious(Too long or slow), Monotonous(same)  
            c. Depends on man power  

    2. Automation - Testing a software / application with the help of tools / technology.  
                  - Write script once and run it many times.    
 
                  ~ Advantages - 
                        a. Faster execution
                        b. Time saving
                        c. Less man power required
                        d. Consistency and accuracy
                        e. Reusability
                  ~ Disadvantage-
                        a. Initial cost is high
                        b. Technical issues possible
                        c. Programming skills is required  
                        d. No human judgement
